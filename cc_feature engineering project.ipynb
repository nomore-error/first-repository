```python
# 거래시간 심야시간인지 판단

df["hour"] = df["trans_date_trans_time"].dt.hour
df["is_night"] = df["hour"].apply(lambda x: 1 if (x >= 22 or x <= 5) else 0)
print(df["is_night"].value_counts())
print(df["is_night"].value_counts(normalize=True))
```

```python
# 거주지와 거래장소가 멀리 있는가 판단 (어려워서 그냥 긁어옴)

import numpy as np

# 벡터화된 haversine 함수
def haversine_vec(lat1, lon1, lat2, lon2):
    # degree -> radian
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    R = 6371.0  # 지구 반지름(km)
    return R * c

# 고객 주소 ↔ 거래 장소 간 거리(km)
df["distance_km"] = haversine_vec(
    df["cst_lat"].to_numpy(),
    df["cst_long"].to_numpy(),
    df["merch_lat"].to_numpy(),
    df["merch_long"].to_numpy()
)

# 빠른 확인
print(df["distance_km"].describe())
```

```python
#평균값보다 더 멀리 떨어져있는 것들은 대상으로 잡음
df["far_from_home"] = (df["distance_km"] >= 77).astype(int)
print(df["far_from_home"].value_counts(normalize=True))
```
