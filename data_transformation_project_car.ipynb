```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
s_merged = ss.fit_transform(merged_df[['Price','Registration_Year','Engine']])
s_merged_df = pd.DataFrame(s_merged, columns=['Price','Registration_Year','Engine'])

s_merged_df
s_merged_df.isna().sum()
s_merged_df = s_merged_df[s_merged_df.isna().sum(axis=1) <= 0]
```

```python
from sklearn.preprocessing import RobustScaler

rs = RobustScaler()
r_merged = rs.fit_transform(merged_df[['Price','Mileage(miles)']])
r_merged_df = pd.DataFrame(r_merged, columns=['Price','Mileage(miles)'])

r_merged_df
r_merged_df.isna().sum()
```

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

pca = PCA()
pca.fit(s_merged_df)

cumsum = np.cumsum(pca.explained_variance_ratio_)
n_components_70 = np.argmax(cumsum >=0.7) +1
print(n_components_70)
print(pca.components_)
```
