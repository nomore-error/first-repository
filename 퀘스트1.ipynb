```python
#메인퀘스트1번

import random

class Account :
    bank_name = "SC은행"
    account_count = 0
    
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
        self.account_number = self.generate_account_number()
        self.deposit_count=0
        self.deposit_history_list = []
        self.withdrawal_history_list=[]
        Account.account_count += 1
        
        print(f"{self.owner}님의 계좌가 생성되었습니다. 계좌번호: {self.account_number}")
        
    def generate_account_number(self):
        part1 = str(random.randint(100,999))
        part2 = str(random.randint(10,99))
        part3 = str(random.randint(100000,999999))
        return f"{part1}-{part2}-{part3}"
    
    def get_account_num(self):
        return Account.account_count
    
    def deposit(self, amount):
        if amount >=1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_history_list.append(amount)
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.balance:,}원")
            
            if self.deposit_count %5 ==0:
                interest = int(self.balance*0.01)
                self.balance += interest
                print(f"이자 {interest}원이 추가 지급되었습니다.")
        else:
            print("입금은 1원 이상만 가능")
            
            
    def withdraw(self,amount):
        if amount > self.balance:
            print("잔액부족")
        else:
            self.balance -= amount
            self.withdrawal_history_list.append(amount)
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance:,}원")
            
    def display_info(self):
        formatted_balance = f"{self.balance:,}"
        print(f"은행이름: {Account.bank_name}, 예금주: {self.owner}, "
              f"계좌번호: {self.account_number}, 잔고: {formatted_balance}원"
        )

            
    def deposit_history(self):
        print(f"{self.owner}님의 입금내역: {self.deposit_history_list}")
        
        
    def withdrawal_history(self):
        print(f"{self.owner}님의 출금내역: {self.withdrawal_history_list}")

        
#Q8번 

acc1 = Account("최현석", 500000)
acc2 = Account("윤남노", 2000000)
acc3 = Account("김풍", 10000)

account_list = [acc1,acc2,acc3]

#Q9번

for account in account_list:
    if account.balance >= 1000000:
        account.display_info()
        
    
```
